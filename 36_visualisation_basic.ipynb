```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('2019-01-01', periods=1000))
```

```python
ts1 = ts.cumsum()
ts1.plot()
```

```python
ts2 = ts1.cummin()
ts2.plot()
```

```python
ts3 = ts1.cummax()
ts3.plot()
```
