```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creating dataset, uniform
df = pd.DataFrame(np.random.rand(10000))

# histogram
df.hist(bins=100)

df.plot(kind='hist', bins=100)
```

```python
# creating dataset, normal
df1 = pd.DataFrame(np.random.randn(10000))
df1.hist(bins=50, color='red', alpha=0.5)
```

```python
# multiple histogram
df = pd.DataFrame({'mu_0_sigma_1': np.random.randn(10000),
                   'mu_1_sigma_2': 2 * np.random.randn(10000) + 1,
                   'mu_8_sigma_3': 3 * np.random.randn(10000) + 8})
df.plot.hist(bins=100, cmap='viridis', alpha=0.5,
             title='Different Normal Distribution')
```

```python
# cumulative plot
df['mu_8_sigma_3'].plot.hist(cumulative=True, bins=1000, color='green',
                             alpha=0.5)
```

```python
# share screen for plots
df.hist(sharex=True, sharey=True)
```
