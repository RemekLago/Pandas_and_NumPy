```python
import pandas as pd
import numpy as np

index = pd.date_range('01-01-2020', periods=10_000)
df = pd.DataFrame(np.random.randn(10_000), index=index)
```

```python
df_cumsum = df.cumsum()
df_cumsum.plot(kind="line")
```
