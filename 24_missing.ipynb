```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(10, 4),
                columns=['one', 'two', 'three', 'four'])
```

```python
for row in df.values:
    switch = np.random.choice([0, 1])
    if switch:
        i = np.random.choice([0, 1, 2, 3])
        row[i] = np.nan
```

```python
# isnull()
df.isnull()

df['one'].isnull()

df[df['one'].isnull()]
df[~df['one'].isnull()]
```

```python
# notnull()
df.notnull()
df[df.notnull()]

df['one'].notnull()
```
